1) 20 times irão disputar um campeonato de futebol e no total cada time disputa 36 partidas.Assumindo que a probabilidade de um time vencer uma partida é de 1/3,qual a probabilidade de um time que disputa esse campeonato, terminar o campeonato com 10 vitórias?
Dica:use a distribuição binomial.

P(X = k) = (n!/(k!x(n-k)!))x(p^k)x(1-p)^(n-k)

n é o número total de tentativas,

k é o número de sucessos desejados

p é a probabilidade de sucesso em uma única tentativa.

```python
#biblioteca para calcular o fatorial
import math
print("Cada partida: Probabilidade de vitória = 1/3")
print("n = 36,k = 10 e p = 1/3")
n = 36
k = 10
p = 1/3

combinacao = (math.factorial(n))/((math.factorial(k))*(math.factorial(n-k)))

probabilidade = combinacao*pow(p,k)*(pow((1-p),(n-k)))

print(f'Combinaçao: {combinacao}')
print(f'Probabilidade do time obter 10 vitorias no campeonato: {probabilidade}')
```

3)José lança um dado de 6 faces,não viciado,12 vezez.Qual a probabilidade de que a face de valor 3 caia virada para cima duas vezes consecutivas?

```python
#pegando a probabilidade de sairem dois 3 de forma consecutiva com dois lançamentos
possibilidades = []
for caso1 in list(range(1,7)):
  for caso2 in list(range(1,7)):
    poss = [caso1,caso2]
    possibilidades.append(poss)
print('Todas as possibilidades em dois lançamentos consecutivos')
print(possibilidades)
print(f'\nProbabilidade em dois lançamentos consecutivos:{1/len(possibilidades)}')

#contando a quantidade de lançamentos consecutivos em 12 lancamentos
lanc_con = []
for lanc1 in list(range(1,13)):
  for lanc2 in list(range(1,13)):
    if ((lanc2 - lanc1) == 1):
      lanc_con.append([lanc1,lanc2])

print(f'\nTotal de lançamentos consecutivos:{len(lanc_con)}')

print(f'\nProbabilidade de sair 3 em dois lancçamentos consecutivos:{(1/len(possibilidades))*len(lanc_con)}')
```

4) Em um programa de televisão existe uma dinãmica com vários desafios, um desses desafios é chamado de adivinhar a soma,em que o participante tem que dizer um valor e esse valor tem que ser a soma dos valores de duas bolas que ele tira aleatoriamente de uma caixa,que contém 30 bolas numeradas de 1 a 30.Qual é a probabilidade de que um participante tire duas bolas cuja soma seja 24?    

```python
#pegando a probabilidade de ao pegar daus bolas aleatorias a soma ser 24
possibilidades = []
for bola1 in list(range(1,31)):
  for bola2 in list(range(1,31)):
    soma = bola1+bola2
    possibilidades.append(soma)
print('Todas as possibilidades de soma')
print(possibilidades)

#calculando a quantidade de vezes que a soma da 24 entre todas as possibilidade
chances = 0;
for desejado  in possibilidades:
  if desejado == 24:
    chances += 1;

print(f'\nTotal de possibilidades de ser 24: {chances}')
print(f'\nProbabilidade de ser 24: {chances/len(possibilidades)}')
```

7)Em uma cidade de colonização alemã, a probabilidade de uma pessoa falar alemão é de 60%. Selecionando-se ao acaso 4 pessoas desta cidade,qual a probabilidade de, exatamente, 3 delas não falarem alemão em valores percentuais?
Dica: use distribuiçao binomial


```python
import math
# numero de pessoas selecionadas
n = 4
# numero de pessoas que nao falam alemao
k = 3
# probabilidade de nao falar alemao
p = 0.4

#P(X = k) = n!/(k!(n-k)!) * (p)^k * (1-p)^(n-k)

probabilidade = (math.factorial(n)/(math.factorial(k)*math.factorial(n-k))) * (p**k) * (1-p)**(n-k)

print(f'Probabilidade = {round(probabilidade*100,2)}%')
```

10)Considere seguinte a funçao densidade de probabilidade v.a.c X ,em que k = 3/2:
image.png

Calcule a media e variancia da variavel aleatoria X.

```python
from scipy import integrate
#calculando a media

media,erro = integrate.quad(lambda x:x*((3/2)*(2*x - x**2)), 0, 1)
print(f'Media = {media}')

#calculando a variancia
# integral de -infinito a +infinito de (x-media)^2 * f(x)
variancia,erro = integrate.quad(lambda x:((x-media)**2)*((3/2)*(2*x - x**2)), 0, 1)
print(f'Variancia = {variancia}')
```
