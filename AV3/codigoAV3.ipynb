## *Analise Descritiva dos Dados*

```python
feridosInicial = data['feridos'].values
feridosInicial
```

```python
diaDaSemana = data['dia_semana'].values
diaDaSemana
```

Fazendo contagem de acidentes por dia da semana

```python
acidentes = [0,0,0,0,0,0,0]


for dia in diaDaSemana:
    if(dia == "segunda"):
        acidentes[0]+=1
    if(dia == "terça"):
        acidentes[1]+=1
    if(dia == "quarta"):
        acidentes[2]+=1
    if(dia == "quinta"):
        acidentes[3]+=1
    if(dia == "sexta"):
        acidentes[4]+=1
    if(dia == "sábado"):
        acidentes[5]+=1
    if(dia == "domingo"):
        acidentes[6]+=1


total=0
i=0
while(i<7):
    total+=acidentes[i]
    i+=1

total

```

fazendo contagem do numero de feridos por dia da seman


```python
feridos=[0,0,0,0,0,0,0]


i=0
while(i < total):
    if(diaDaSemana[i] == "segunda"):
        feridos[0]+=feridosInicial[i]
    if(diaDaSemana[i]  == "terça"):
        feridos[1]+=feridosInicial[i]
    if(diaDaSemana[i]  == "quarta"):
        feridos[2]+=feridosInicial[i]
    if(diaDaSemana[i]  == "quinta"):
        feridos[3]+=feridosInicial[i]
    if(diaDaSemana[i]  == "sexta"):
        feridos[4]+=feridosInicial[i]
    if(diaDaSemana[i]  == "sábado"):
        feridos[5]+=feridosInicial[i]
    if(diaDaSemana[i]  == "domingo"):
        feridos[6]+=feridosInicial[i]
    i+=1
feridos
```

fazendo contagem do numero de feridos graves por dia da semana

```python
feridos_graves=[0,0,0,0,0,0,0]

i=0
while(i < total):
    if(diaDaSemana[i] == "segunda"):
        feridos_graves[0]+=feridos_gravesI[i]
    if(diaDaSemana[i]  == "terça"):
        feridos_graves[1]+=feridos_gravesI[i]
    if(diaDaSemana[i]  == "quarta"):
        feridos_graves[2]+=feridos_gravesI[i]
    if(diaDaSemana[i]  == "quinta"):
        feridos_graves[3]+=feridos_gravesI[i]
    if(diaDaSemana[i]  == "sexta"):
        feridos_graves[4]+=feridos_gravesI[i]
    if(diaDaSemana[i]  == "sábado"):
        feridos_graves[5]+=feridos_gravesI[i]
    if(diaDaSemana[i]  == "domingo"):
        feridos_graves[6]+=feridos_gravesI[i]
    i+=1
feridos_graves
```

organizandos os dados coletados em uma lista

```python
i=0
dia = ['Segunda','Terça','Quarta','Quinta','Sexta','Sábado','Domingo']
dados=[0,0,0,0,0,0,0]
while i < 7:
    dados[i] =[dia[i],acidentes[i],feridos[i],feridos_graves[i]]
    i+=1
dados
```

## Estimação com os dados

Estimações sobre a variavel Feridos e Feridos_Graves

Pegando uma parte de 40% do dataframe "data_coletados" para feridos

```python
# Pegando uma amostra aleatória de 40% do banco de dados para Feridos
amostraF = data_coletados.sample(frac=0.4)

# Imprimindo a amostra
print(amostraF)
```

Pegando uma parte de 40% do dataframe "data_coletados" para feridos_graves

```python
# Pegando uma amostra aleatória de 40% do banco de dados para FeridosGraves
amostraFG = data_coletados.sample(frac=0.4)

# Imprimindo a amostra
print(amostraFG)
```

media amostral

```python
# Calculando a média amostral de Feridos e Feridos_Graves
m_amostralF = amostraF['Feridos'].mean()
m_amostralFG = amostraFG['Feridos_Graves'].mean()

# Imprimindo a média amostral
print('Média amostral de Feridos: {:.2f}'.format(m_amostralF))

print('Média amostral de Feridos_Graves: {:.2f}'.format(m_amostralFG))

print("As medias ficam em torno desses valores")

```

Desvio padrao amostral

```python
# Calculando o desvio padrão da coluna Feridos e Feridos_Graves
desvio_padraoF = np.std(data_coletados['Feridos'], ddof=1)
desvio_padraoFG = np.std(data_coletados['Feridos_Graves'], ddof=1)

# Imprimindo o desvio padrão
print('Desvio padrão Feridos :{:.2f}'.format(desvio_padraoF))
print('Desvio padrão Feridos_Graves:{:.2f}'.format(desvio_padraoFG))

print("\nDesvios altos ,geram uma variacao significativa da media amostral")
```

Variancia amostral

```python
# Calculando a variância de Ocorrencias
varianciaF = np.var(data_coletados['Feridos'], ddof=1)
varianciaFG = np.var(data_coletados['Feridos_Graves'], ddof=1)

# Imprimindo a variância
print('Variância de feridos: {:.2f}'.format(varianciaF))
print('Variância de feridos_graves: {:.2f}'.format(varianciaFG))
```

Intervalo de Confiança da Media

```python
from scipy.stats import norm
# Definindo o nível de confiança e o tamanho das amostras
nivel_confianca = 0.95
n1 = len(amostraF)
n2 = len(amostraFG)

# Calculando o erro padrão das médias
erro_padraoF = desvio_padraoF / np.sqrt(n1)
erro_padraoFG = desvio_padraoFG / np.sqrt(n2)

# Calculando o valor crítico da distribuição normal
valor_critico = norm.ppf((1 + nivel_confianca) / 2)

# Calculando o intervalo de confiança
limite_inferiorF = m_amostralF - valor_critico * erro_padraoF
limite_superiorF = m_amostralF + valor_critico * erro_padraoF

limite_inferiorFG = m_amostralFG - valor_critico * erro_padraoFG
limite_superiorFG = m_amostralFG + valor_critico * erro_padraoFG

# Imprimindo o intervalo de confiança
print('O intervalo de confiança de 95% para a média de Feridos: {:.2f} - {:.2f}'.format(limite_inferiorF, limite_superiorF))
print('O intervalo de confiança de 95% para a média de Feridos_Graves: {:.2f} - {:.2f}'.format(limite_inferiorFG, limite_superiorFG))
```

Com 95% de confiança, podemos afirmar que as médias estão dentro desses intervalos.

Nesse caso, o intervalo de confiança sugere que as médias provavelmente estarão próximas às médias amostrais, mas poderão ser um pouco mais baixas ou mais altas.

## Testes de Hipoteses

Trabalhamos com duas hipoteses

A fim de sabermos quais hipóteses serão rejeitadas, será utilizado o teste t de Student independente, que retorna o p-valor e a estatística do teste. E, adotando o valor de significância padrão de 0.05 ou 5%, caso o p-valor retornado da função ttest_ind seja maior ou igual ao valor de significância, a hipótese nula é aceita e a hipótese alternativa é rejeitada, caso contrário, a hipótese nula é rejeitada e a hipótese alternativa é aceita.

O primeiro teste contém as seguintes hipóteses:

Hipótese nula: a média e desvio padrão de Feridos na primeira metade das semanas são maiores ou iguais a segunda metade das semanas.
Hipótese alternativa: a média e desvio padrão de Feridos na primeira metade das semanas é são menores que a segunda metade das semanas.

```python
import statistics
#Primeiro teste

metadeP = data_coletados['Feridos'].iloc[0:3]
metadeS = data_coletados['Feridos'].iloc[3:6]




print("A Média da qtde de Feridos durante a primeira metade das semanas : {0}".format(round(metadeP.mean(),2)))
print("A Média da qtde de Feridos durante a segunda metade das semanas : {0}".format(round(metadeS.mean(),2)))

print("O Desvio Padrão da qtde de Feridos durante a primeira metade das semanas: {0}".format(round(statistics.pstdev(metadeP),2)))
print("O Desvio Padrão da qtde de Feridos durante a segunda metade das semanas: {0}".format(round(statistics.pstdev(metadeS),2)))

estat_ttest, pvalor = scipy.stats.ttest_ind(metadeP, metadeS)
print("A estatística do Teste é {0} e o p-valor é {1}".format(estat_ttest, pvalor))

if pvalor < 0.05:
    print("Rejeitamos a Hipótese Nula")
else:
    print("Não Rejeitamos a Hipótese Nula")
```

O segundo teste contém as seguintes hipóteses:

Hipótese nula: a média e desvio padrão de Feridos_Graves na primeira metade das semanas são maiores ou iguais a segunda metade das semanas. Hipótese alternativa: a média e desvio padrão de Feridos_Graves na primeira metade das semanas é são menores que a segunda metade das semanas.

```python
import statistics
#Primeiro teste

metadeP = data_coletados['Feridos_Graves'].iloc[0:3]
metadeS = data_coletados['Feridos_Graves'].iloc[3:6]




print("A Média da qtde de Feridos gravemente durante a primeira metade das semanas : {0}".format(round(metadeP.mean(),2)))
print("A Média da qtde de Feridos gravemente durante a segunda metade das semanas : {0}".format(round(metadeS.mean(),2)))

print("O Desvio Padrão da qtde de Feridos gravemente durante a primeira metade das semanas: {0}".format(round(statistics.pstdev(metadeP),2)))
print("O Desvio Padrão da qtde de Feridos gravemente durante a segunda metade das semanas: {0}".format(round(statistics.pstdev(metadeS),2)))

estat_ttest, pvalor = scipy.stats.ttest_ind(metadeP, metadeS)
print("A estatística do Teste é {0} e o p-valor é {1}".format(estat_ttest, pvalor))

if pvalor < 0.05:
    print("Rejeitamos a Hipótese Nula")
else:
    print("Não Rejeitamos a Hipótese Nula")
```

Todos os calculos foram feitos utilizando o teste do t student e em ambos os testes a hipotese nula nao foi rejeitada pois o valor de significancia padrão 5% não foi ultrapassado.

## Regressão Linear

 modelar a relação entre uma variável independente (ou variáveis independentes) e uma variável dependente, de modo a prever ou explicar a variabilidade na variável dependente com base nas variáveis independentes

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from IPython.display import display, Markdown, Latex
```

```python
data_copia = data_coletados.copy()
diaS={
    'Segunda':1,
    'Terça':2,
    'Quarta':3,
    'Quinta':4,
    'Sexta':5,
    'Sábado':6,
    'Domingo':7

}

# faz a substituição no dataframe
data_copia['Dia'] = data_copia['Dia'].replace(diaS)
```

Analisando as regressão com as variáveis Dia e Feridos.

```python
# Realizando a análise de regressão com os dados de 2022
# (substitua 'x' e 'y' pelos nomes das colunas que contêm as variáveis independentes e dependentes, respectivamente)
regressao = LinearRegression()

X = data_copia[['Dia']]
Y = data_copia['Feridos']

regressao.fit(X, Y)

coef_angular = regressao.coef_[0]
coef_linear = regressao.intercept_

# Plotando o resultado
import matplotlib.pyplot as plt

plt.scatter(X.iloc[:,0], Y)
plt.plot(X.iloc[:,0], regressao.predict(X), color='red')
plt.xlabel('Dias')
plt.ylabel('Feridos')
plt.show()

# Calculando o R-squared
print(regressao.score(X,Y))

print(f"A equação de regressão é: Y = {coef_angular:.2f} * X1 + {coef_linear:.2f}")
```
