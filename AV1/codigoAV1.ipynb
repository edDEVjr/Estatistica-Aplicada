Criando um  novo dataframe apenas com as estatisticas do Flamengo

```python
data_Fla = data.query('team == "Flamengo"')
```

```python
data_Fla
```

Adiconando temporadas,posicoes,pontos e quantidade de gols marcados do Flamengo de 2007 a 2022 em variaveis

```python
anos = data_Fla['season'].values
anos
```

```python
posicoes = data_Fla['place'].values
posicoes
```

```python
pontos = data_Fla['points'].values
pontos
```

```python
gols_mar = data_Fla['goals_for'].values
gols_mar
```

Listas com anos em relacao aos outros dados que serao utilizados 

```python
ano_posicao = []
for x in range(4,20):
    ap = [anos[x],posicoes[x]]
    ano_posicao.append(ap)
print(ano_posicao)
    
```

```python
ano_pontos = []
for x in range(4,20):
    ap = [anos[x],pontos[x]]
    ano_pontos.append(ap)
print(ano_pontos)
```

```python
ano_gols = []
for x in range(4,20):
    ag = [anos[x],gols_mar[x]]
    ano_gols.append(ag)
print(ano_gols)
```

Criando Dataframes com as listas anteriores

```python
posicaoData = pd.DataFrame(ano_posicao,columns=['Ano','Posicao'])
posicaoData
```

```python
pontosData = pd.DataFrame(ano_pontos,columns=['Ano','Pontos'])
pontosData
```

```python

golsData = pd.DataFrame(ano_gols,columns=['Ano','Gols Marcados'])
golsData
```

Media ,Moda, Mediana e Desvio Padrao

```python
media_posicao = posicaoData['Posicao'].mean()
media_pontos = pontosData['Pontos'].mean()
media_gols = golsData['Gols Marcados'].mean()

median_posicao = posicaoData['Posicao'].median()
median_pontos = pontosData['Pontos'].median()
median_gols = golsData['Gols Marcados'].median()

mdp =list(posicaoData['Posicao'].mode())
mdpo = list(pontosData['Pontos'].mode())
mdg = list(golsData['Gols Marcados'].mode())

moda_posicao = mdp[0]
moda_pontos = mdpo[0]
moda_gols = mdg[0]



dp_posicao = posicaoData['Posicao'].std()
dp_pontos = pontosData['Pontos'].std()
dp_gols = golsData['Gols Marcados'].std()



```

```python
print(f"Media de Posição:{media_posicao}")
print(f'Media de Pontos:{media_pontos}')
print(f'Media de Gols:{media_gols}')
```

```python
print(f"Mediana de Posição:{median_posicao}")
print(f'Mediana de Pontos:{median_pontos}')
print(f'Mediana de Gols:{median_gols}')
```

```python
print(f"Moda de Posição:{moda_posicao}")
print(f'Moda de Pontos:{moda_pontos}')
print(f'Moda de Gols:{moda_gols}')
```

```python
print(f"Desvio Padrao de Posição:{dp_posicao}")
print(f'Desvio Padrao de Pontos:{dp_pontos}')
print(f'Desvio Padrao de Gols:.{dp_gols}')
```

Coeficiente de Variacao CV=dp/Media


```python
CVPosicao = 4.979959839195493/6.0
CVPontos = 12.159358535712318/61.625
CVGols = 12.495332461896854/56.0
```

```python
print('Coficiente de Variacao da Posição: '+str(CVPosicao))
print('Coficiente de Variacao de Pontos: '+str(CVPontos))
print('Coficiente de Variacao de Gols: '+str(CVGols))
```

Criando um dataFrame com valores encontrados anteriormente

```python
valoresEsta = pd.DataFrame({
    'Dado':['Posicao','Pontos','Gols'],
    'Media': [media_posicao,media_pontos,media_gols],
    'Moda': [moda_posicao,moda_pontos,moda_gols],
    'Mediana': [median_posicao,median_pontos,median_gols],
    'Desvio Padrao': [dp_posicao,dp_pontos,dp_gols],
    'Coef. Variacao': [CVPosicao,CVPontos,CVGols]
})
valoresEsta
```

Gerando Graficos com matplotlib

Grafico de Posicao por ano

Grafico de Gols Marcados por Ano

```python
plt.bar(golsData['Ano'],golsData['Gols Marcados'])
plt.title('Gols do Flamengo por Ano no Brasileirao')
plt.ylabel('Gols')
plt.xlabel('Ano')
```

Grfico em barra comparando gols e pontos

```python
plt.bar(pontosData['Ano'],pontosData['Pontos'],color='black')
plt.bar(golsData['Ano'],golsData['Gols Marcados'],color='red')
plt.title('Pontos/Gols do Flamengo por Ano no Brasileirao')
plt.ylabel('Pontos(Preto)/Gols(Vermelho)')
plt.xlabel('Ano')
```

Grafico de linha de Gols do Flamengo no Brasileirao

```python
plt.plot(golsData['Ano'],golsData['Gols Marcados'] , marker='o', linestyle='-', color='b')
plt.xlabel('Ano')
plt.ylabel('Gols')
plt.title('Gols do Flamengo no Brasileirao')
plt.grid(True)
plt.show()
```
